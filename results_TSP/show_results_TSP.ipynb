```python
import matplotlib.pyplot as plt
from collections import defaultdict

    
```

```python
def Average(lst):
    return sum(lst) / len(lst)

def ret_dict_from_file(path):
    file1 = open(path, 'r')
    Lines = file1.readlines()
    mydict = {}  
    for line in Lines:
        x = line.split(": ")
        key =  int(x[0])
        value = float(x[1].rstrip("\n"))
        
        if key in mydict:
            mydict[key].append(value)
        else:
           
            mydict[key] = [value]
            
    for key in mydict:
        mydict[key] = Average(mydict[key])
    
    keys_list = list(mydict.keys())
    values_list = list(mydict.values())
    
    
    return keys_list , values_list

```

```python
x1, y1 = ret_dict_from_file("length_path_ACO_TSP.txt")
# plotting the line 1 points
plt.plot(x1, y1, label = "line 1")
# plt.plot(keys_list, values_list)
 
# naming the x axis
plt.xlabel('Number of cities')
# naming the y axis
plt.ylabel('Track length')
 
# giving a title to my graph
plt.title('Track length in TSP problem in ACO algorithm')
 
# function to show the plot
plt.show()
```
